### **3.2 Continuity and Differentiation**

Let f: D→R be a real-valued function on D⊆R^d. Then f is said to be constinuous at a∈D if lim_{x->a} f(x) = f(a).

Let f: D→R where D⊆R and let x_0∈D be an interior point of D. The derivative of f at x_0 is f'(x_0)=df(x_0)/dx=lim_{h→0}(f(x_0+h)-f(x_0))/h, provided the limit exists.

```python
from sympy import Symbol, limit, oo

x = Symbol('x')
y = x**2

limit(y, x, oo)
```

### **3.3 Unconstrained Optimization**

Let f:R^d→R. The point x*∈R^d is a local minimizer of f over R^d if there is δ>0 such that f(x)≽f(x*), ∀x∈B_δ(x*)\{x*}. If the inequality is strict, we say that x* is a strict local minimizer.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(0,8)
y = x**2
x2 = np.linspace(0,6)
y2 = x*4.5

plt.plot(x,y,'b', x2,y2,'r') 
plt.title('Unconstrained Optimization')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
